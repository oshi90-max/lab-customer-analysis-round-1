```python
import pandas as pd
import numpy as np

file1 = pd.read_csv('file1.csv')
file2 = pd.read_csv('file2.csv')
file3 = pd.read_csv('file3.csv')


column_names1 = file1.columns
column_names2 = file2.columns
column_names3 = file3.columns

#display(column_names1)
#display(column_names2)
#display(column_names3)

file3 = file3.rename(columns={'State':'ST'})
display(column_names3)

cols1 = []
for i in range (len(file1.columns)):
    cols1.append(file1.columns[i].lower())
file1.columns = cols1

print (file1.columns)

cols2 = []
for i in range (len(file2.columns)):
    cols2.append(file2.columns[i].lower())
file2.columns = cols2

print (file2.columns)

cols3 = []
for i in range (len(file3.columns)):
    cols3.append(file3.columns[i].lower())
file3.columns = cols3

print (file3.columns)

file1['customer lifetime value'] = file1['customer lifetime value'].str.replace('%', '')
file2['customer lifetime value'] = file2['customer lifetime value'].str.replace('%', '')
file1['customer lifetime value'] = pd.to_numeric(file1['customer lifetime value'], errors='coerce')
file2['customer lifetime value'] = pd.to_numeric(file2['customer lifetime value'], errors='coerce')
file3['customer lifetime value'] = pd.to_numeric(file3['customer lifetime value'], errors='coerce')
file1['customer lifetime value'] = file1['customer lifetime value']/100
file2['customer lifetime value'] = file1['customer lifetime value']/100


data = pd.concat([file1, file2, file3], axis=0) 
data.shape
data
```
